```python
import pandas as pd
import os.path
root_path = os.path.dirname(os.getcwd())

# import food inspection data
inspections = pd.read_csv(os.path.join(root_path, "DATA/food_inspections.csv"))
```

```python
columns = [("v_" + str(num)) for num in range(1, 46)]
columns.append("v_75")
```

```python
def split_violations(violations):
    values = pd.Series([])

    violations = violations.split(' | ')
    for violation in violations:
        index = "v_" + violation.split('.')[0]
        values[index] = 1
    return values
```

```python
values_df = inspections.head().violations.apply(split_violations)
```

```python
violation_matrix = pd.DataFrame(values_df, columns=columns).fillna(0)
```
