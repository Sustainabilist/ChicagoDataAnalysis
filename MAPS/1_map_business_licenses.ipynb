```python
# Install folium
!pip install folium

# Import necessary packages 
import folium
from folium import plugins
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

#### Import and clean data

```python
# Import business license data
import os.path
root_path = os.path.dirname(os.getcwd())
business_licenses  = pd.read_csv(os.path.join(root_path,"DATA/business_licenses.csv"))
```

```python
business_licenses = business_licenses.dropna()
# business_licenses.shape returns rows and columns
number_of_licenses = len(business_licenses)
```

#### Map licenses

```python
# Use folium to visualize map
m = folium.Map([41.8600, -87.6298], zoom_start=10)
m
```
